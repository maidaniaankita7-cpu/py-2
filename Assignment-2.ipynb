```python
#Q.3 What are missing values in Pandas and how are they represented?
#Create a DataFrame with missing values and write code to:

#• Detect missing values
#• Replace them with the column mean

#Explain each step.

import pandas as pd
import numpy as np

data = {
    "Marks":[80, np.nan, 60, 90],
    "Age":[20,22,np.nan,21]
}

df = pd.DataFrame(data)

# detect missing
print(df.isnull())

# replace with mean
df["Marks"].fillna(df["Marks"].mean(), inplace=True)
df["Age"].fillna(df["Age"].mean(), inplace=True)

print(df)
```

```python
#Q.4 Explain boolean indexing in NumPy or Pandas.
#Create a DataFrame with at least 5 columns and filter rows based on two conditions using logical operators. Explain how filtering works internally.
data = {
    "Name":["A","B","C","D"],
    "Age":[20,25,30,22],
    "Marks":[80,55,90,40],
    "City":["A","B","A","C"],
    "Dept":["IT","HR","IT","Sales"]
}

df = pd.DataFrame(data)

filtered = df[(df["Age"]>21) & (df["Marks"]>50)]
print(filtered)
```

```python
#Q.5 What is the purpose of the groupby() function in Pandas?
#Create a DataFrame with categorical data (e.g., department & salary) and calculate the average salary per department using groupby(). Explain the output.
data = {
    "Dept":["IT","HR","IT","HR","Sales"],
    "Salary":[50000,40000,60000,45000,30000]
}

df = pd.DataFrame(data)

avg_salary = df.groupby("Dept")["Salary"].mean()
print(avg_salary)
```
